1. Investigue el uso o propósito de las siguientes carpetas en un sistema operativo basado en el kernel Linux: <br>
a. Bin: Contiene archivos binarios ejecutables. Estos archivos son comandos básicos que se utilizan tanto por el administrador como por los usuarios. <br>
b. <b>Opt</b>: Contiene paquetes opcionales de terceros, que se instalan por fuera la configuración base del sistema.<br>
c.	<b>Boot</b>: Contiene todos los archivos necesarios para arrancar el sistema.<br>
d.	<b>Dev</b>: Directorio que contiene archivos de dispositivos externos, estos archivos no son archivos normales, sino que actúan como nodos que representan dispositivos físicos.<br>
e.	<b>Sbin</b>: Contiene ejecutables binarios no accesible para cualquier usuario. Solo pueden ser usados por el administrador del sistema.<br>
f.	<b>Etc</b>: Contiene los archivos de configuración del sistema. Allí se almacenan las configuraciones de las aplicaciones y servicios instalados.<br>
g.	<b>Home</b>: Sirve como el espacio de almacenamiento principal de los usuarios<br>
h.	<b>Srv</b>: Es donde se almacenan los archivos y directorios relativos a servicios web, FTP, CVS, etc. <br>
i.	<b>Lib</b>: Contiene las bibliotecas compartidas esenciales para el funcionamiento del sistema y de los programas. <br>
j.	<b>Tmp</b>: Es donde se almacenan todos los archivos temporales. <br>
k.	<b>Usr</b>: Contiene archivos ejecutables, documentación bibliotecas y otros datos que son compartidos para todos los usuarios como solo lectura. <br>
l.	<b>Media</b>: se utiliza para el montaje automático de dispositivos extraibles. <br>
m.	<b>Var</b>: Directorio que contiene la información de variables del sistema (logs,spools), caché y archivos temporales que no deben eliminarse al reiniciar. <br>
n.	<b>Mnt</b>: Directorio utilizado como punto de montaje temporal para sistemas de archivos. Los administradores realizan el montaje de manera manual.

Cuando se despliega un modelo de Machine Learning en un entorno productivo, las librerías de Machine Learning o Deep Learning en Python ofrecen diversas herramientas para encapsular todo el flujo necesario para realizar predicciones a partir de datos de entrada. En el caso de scikit-learn, esto se logra a través del concepto de Pipeline (Docs Scikit Learn) Investigue cómo funciona esta herramienta e implemente un ejemplo sencillo que demuestre su uso.

```python
from sklearn.pipeline import Pipeline
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data_dic=load_diabetes()
data=data_dic['data']
feature=data_dic['feature_names']
target=data_dic['target']
data_df=pd.DataFrame(data=data,columns=feature)

X_train, X_test, y_train, y_test = train_test_split(data_df, target, random_state=42)
#Se crean los pasos que ejecutará el pipeline.
steps=[('Scaler',StandardScaler()),('Regresion',LogisticRegression())]
pipe=Pipeline(steps=steps)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
```
